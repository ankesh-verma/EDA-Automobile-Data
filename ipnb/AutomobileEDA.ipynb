## 4. Data Preprocessing <a name="section4"></a> 



#### 4.1.  Data Cleaning<a name="section4.1"></a> 

```python
df_Automobile.head()
```

- <b><mark style='background-color:#EBA4A4'>Fixing Bore Problem (removing '?')</mark></b>

```python
# check Bore Entry
df_Automobile['bore'].unique()
```

```python
df_Automobile['bore'].value_counts()
```

```python
bore_val = df_Automobile[~(df_Automobile['bore'] == '?')]['bore'].median()   # Find Median of bore column
```

```python
df_Automobile['bore'] = df_Automobile['bore'].astype(str).str.replace('?', '3.31').astype(float)  # Replace '?' with Median++
```

```python
df_Automobile['bore'].value_counts()  # Recheck Median Value
```

- <b><mark style='background-color:#EBA4A4'>Fixing Stroke Column Issue ( removing '?' )</mark></b>

```python
df_Automobile['stroke'].value_counts()
```

```python
stroke_val = df_Automobile[~(df_Automobile['stroke'] == '?')]['stroke'].median()   # Find Median of bore column
print("Median of stroke column =>",stroke_val)
```

```python
#replace ? with median
df_Automobile['stroke'] = df_Automobile['stroke'].astype(str).str.replace('?', '3.29').astype(float)
```

```python
df_Automobile['stroke'].value_counts()
```

- <b><mark style='background-color:#EBA4A4'>Fixing 'horsepower' Column Issue ( removing '?' )</mark></b>

```python
df_Automobile.horsepower.value_counts()
```

```python
HP_val = df_Automobile[~(df_Automobile['horsepower'] == '?')]['horsepower'].median()   # Find Median of bore column
print("Median of horsepower column =>",HP_val)
```

```python
hp = str(HP_val) # convert hosepower median into string for replacing '?'
df_Automobile['horsepower'] = df_Automobile['horsepower'].astype(str).str.replace('?', hp).astype(float)
```

```python
#Recheck horsepower value
df_Automobile.horsepower.value_counts()  
```

- <b><mark style='background-color:#EBA4A4'>Fixing 'price' column (removing '?')</mark></b>

```python
df_Automobile.price.value_counts()[:5]   # Print top 5 values
```

```python
df_Automobile['price'].unique()  # check all values of price
```

```python
price_val = df_Automobile[~(df_Automobile['price'] == '?')]['price'].median()   # Find Median of bore column
print("Median of price column =>",int(price_val)) # type casting to int 
```

```python
cost = str(int(price_val))
df_Automobile['price'] = df_Automobile['price'].astype(str).str.replace('?', cost).astype(int)
```

```python
df_Automobile.price.value_counts()[:5]  # '?' replaced with median
```

```python
#drop Column which is not required for EDA
df_Automobile.drop([ 'symboling', 'normalized-losses', 'num-of-doors' ], axis=1, inplace=True)
df_Automobile.head()
```

```python
df_Automobile.price.unique()
```

```python
df_Automobile.isnull().sum()
```
